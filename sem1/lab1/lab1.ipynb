```python
import numpy as np
```

```python
def density_mean(mass_mean, a_mean):
    return 1000 * mass_mean / a_mean**3
```

```python
def density_error(mass_erorr, mass_mean, a_error, a_mean):
    coef = np.sqrt(
        (mass_erorr/mass_mean) ** 2 + 3 * (a_error / a_mean) ** 2
    )
    return density_mean(mass_mean=mass_mean, a_mean=a_mean) * coef, coef
```

```python
a_means = [10.413, 10.428, 10.08, 10.208]
mass_means = [2.703, 8.807, 19.607, 4.59]
a_errors = [0.02, 0.039, 0.017, 0.2]
mass_erorrs = [0.011, 0.011, 0.011, 0.01]

for i in range(4):
    d_e, c = density_error(
        mass_erorr=mass_erorrs[i],
        mass_mean=mass_means[i],
        a_error=a_errors[i],
        a_mean=a_means[i]
    )
    print(f"density_error : {d_e}, coef: {c:3f}, density_mean : {density_mean(mass_mean=mass_means[i], a_mean=a_means[i])}" )
```
